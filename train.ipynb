```python
from monai.networks.nets import UNet
from monai.networks.layers import Norm
from monai.losses import DiceLoss, DiceCELoss


import torch
from preprocess import prepare
from utilities import train
```

```python
data_dir = 'Liver'
model_dir = 'Liver/results' 
data_in = prepare(data_dir, cache=False)
```

```python
device = torch.device("cpu")
model = UNet(
    spatial_dims=3, # Die 3 steht für ein Dreidimensionales Bild
    in_channels=1,# Die Bilder haben bloß einen Channel der entweder 1 oder 0 ist
    out_channels=2,# Der erste channel ist die Pixelwahrscheinlichkeit für den Hintergrund der zweite für den Vordergrund
    channels=(16, 32, 64, 128, 256), # Anzahl der Kerne des jeweiligen Convolutional layers
    strides=(2, 2, 2, 2), # Stride größe für die Maxpoolingschritte
    num_res_units=2, # units bestehen aus jeweils 2 convolutional layern
    norm=Norm.BATCH, #Normalisierte jeden layer nach der Aktivierungsfunktion
).to(device)
```

```python
#loss_function = DiceCELoss(to_onehot_y=True, sigmoid=True, squared_pred=True, ce_weight=calculate_weights(1792651250,2510860).to(device))
loss_function = DiceLoss(to_onehot_y=True, sigmoid=True, squared_pred=True)
optimizer = torch.optim.Adam(model.parameters(), 1e-5, weight_decay=1e-5, amsgrad=True)# parameter: Modell, Lernrate, paramter der die Gewichtung klein, schützt vor überanpassung, Aktiviert verbesserte Adam Variante

if __name__ == '__main__':
    train(model, data_in, loss_function, optimizer, 30, model_dir)
```
